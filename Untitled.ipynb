```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
train_new = pd.DataFrame(imp.fit_transform(train), columns=train.columns)
test_new = pd.DataFrame(imp.fit_transform(test), columns=test.columns)
```

```python
# Cheking for any missing values in train dataset
missing_train_values = train_new.isna().sum()
missing_train_values[missing_train_values > 0 ]
```
