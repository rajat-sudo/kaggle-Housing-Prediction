## Data Cleaning
### 1. Dealing with missing values

```python
# Checking the percentage of null values in train dataset and taking out the values having more than 
# 40 of the data as missing
missing_train = (train.isna().sum())/train.shape[0]
missing_train[missing_train > 0.3]
```

```python
# Checking the percentage of null values in train dataset
missing_test = (test.isna().sum())/test.shape[0]
missing_test[missing_test > 0.3]
```

```python
# As both the columns having missing values over 49% are the same 
# Hence saving it in another another list
dropping_col = missing_test[missing_test > 0.3].index.to_list()
# Also the Id column does not give any information about the house hence dropping it.
dropping_col.append('Id')
train = train.drop(columns=dropping_col)
test = test.drop(columns=dropping_col)
print(f'{dropping_col} Columns Dropped')
```

```python
# Imputing the rest of the missing values with if numerical = median() and categorical = mode()
# i-integer , u-unsigned, f-float , its np.dype.kind - numpy function
train = train.apply(lambda x: x.fillna(x.median()) if x.dtype.kind in 'iuf' else x.fillna(x.mode()))
test = test.apply(lambda x: x.fillna(x.median()) if x.dtype.kind in 'iuf' else x.fillna(x.mode()))
```

```python
# Extracting all the numerical values in another dataframe
train_num = train.select_dtypes(exclude=['object'])
test_num = test.select_dtypes(exclude=['object'])

# Extracting all categorical values in another dataframe
train_cat = train.select_dtypes(include=['object'])
test_cat = test.select_dtypes(include=['object'])

train_num.head()
```

### Feature Selection for Numerical Columns

```python
# Using Corelation
plt.figure(figsize=(8,10))
num_corr = train_num.corrwith(train_num['SalePrice'],0).sort_values(ascending=False)
num_corr = num_corr.iloc[1:] 
num_corr.plot.barh()
plt.show()
```

```python
# Selecting values with corration of more than 0.4 in the positive direction
num_corr_columns = num_corr[num_corr > 0.2].index
```

```python
train_num[num_corr_columns].head()
```

```python
# Using those Statistically significant columns for PCA
Xt = train_num[num_corr_columns] # train

yt = train_num['SalePrice'] # train
Xtt = test_num[num_corr_columns]# test ( as the test column does not contain the SalePrice Column)
```

```python
# Standardizing the train and test numerical values for PCA

from sklearn.preprocessing import StandardScaler
sc_Xtrain = StandardScaler()
sc_Xtest = StandardScaler()

Xt = sc_Xtrain.fit_transform(Xt)
Xtt = sc_Xtest.fit_transform(Xtt)

```

```python
# Applying PCA on both train and test
from sklearn.decomposition import PCA
pca_train = PCA(n_components=3)
pca_test = PCA(n_components=3)

Xt_pca = pca_train.fit_transform(Xt)
Xtt_pca = pca_test.fit_transform(Xtt)
```

```python
Xt_pca[0:5]
```

### Dimetionality 

```python
# Taking import variables Feature selection using Random forest for categorical variables
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=300 ,random_state=1, max_depth=10)
# One hot encoding for train and test
train_cat=pd.get_dummies(train_cat)
test_cat=pd.get_dummies(test_cat)

rf.fit(train_cat,train['SalePrice'])
features = train_cat.columns
importances = rf.feature_importances_
indices = np.argsort(importances)[-20:]  # top 20 features
```

```python
# Plotting the 
plt.style.use('ggplot')
plt.figure(figsize=(8,10))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
# After seeing the bar graph extracting the top 50 features and reversing it the important features are in
# the last
indices_final = (list(np.argsort(importances)[-43:]))
indices_final.reverse()
# Extracting the names of the top seven columns 
final_cat_col = []
for i in indices_final:
    final_cat_col.append(features[i])
final_cat_col
# Creating a new dataframe with 3 PCA column for train, test

PCA_df_train = pd.DataFrame(Xt_pca, columns=['col1','col2','col3'])
PCA_df_test = pd.DataFrame(Xtt_pca, columns=['col1','col2','col3'])
final_train = pd.concat([PCA_df_train, train_cat[final_cat_col] ], 1)
final_test = pd.concat([PCA_df_test, test_cat[final_cat_col] ], 1)
print('done')
```
